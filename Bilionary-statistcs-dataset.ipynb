```python
lista_bilionarios = list(table['personName'])
display(lista_bilionarios)
```

```python
fortunas = {}
fortunas_n_encontradas = {}

for bilionario in lista_bilionarios:
    bilionario_query = f"Fortuna+{bilionario}"
    URL = f"https://www.google.com/search?q={bilionario_query}"

    headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'}
    
    try:
        page = requests.get(URL, headers=headers)
        page.raise_for_status()
    except requests.exceptions.RequestException as e:
        print(f"Erro ao acessar {URL}: {e}")
        continue

    soup = BeautifulSoup(page.content, 'html.parser')
    fortuna_element = soup.find('div', {'class': 'BNeawe iBp4i AP7Wnd'})
    
    if fortuna_element:
        fortuna = fortuna_element.text
        fortunas[bilionario] = fortuna
        print(f'Fortuna de {bilionario} é de {fortuna}!')
    else:
        print(f"Elemento não encontrado para {bilionario}")
        fortunas_n_encontradas[bilionario] = 'Não encontrado'
```
